### Step 7. View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline

```python
discipline=euro12['Team']+euro12['Yellow Cards']+euro12['Red Cards']
discipline
```

### Step 8. Sort the teams by Red Cards, then to Yellow Cards

```python
euro12.sort_values(by=['Red Cards'], inplace=True)
euro12.sort_values(by=['Yellow Cards'], inplace=True)
```

### Step 9. Calculate the mean Yellow Cards given per Team

```python
euro12['Yellow Cards'].mean()
```

### Step 10. Filter teams that scored more than 6 goals

```python
euro12[euro12['Goals'].apply(lambda x: x>6)]['Team']
```

### Step 11. Select the teams that start with G

```python
euro12[euro12["Team"].apply(lambda x: x.startswith("G"))]
```
